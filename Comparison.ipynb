Description: This notebook uses the save accuracies and training times to compare the frameworks.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
TF_time,TF_acc=np.load('TF+Keras.npy')
PT_time,PT_acc=np.load('PT.npy')
```

```python
plt.figure(figsize=(7,7))
plt.bar([0,1], [TF_acc, PT_acc],color=['blue','red'])
plt.ylabel('Accuracy',fontsize=16)
plt.xticks([0,1],['TensorFlow','PyTorch'],fontsize=15)
```

```python
plt.figure(figsize=(7,7))
plt.bar([0,1], [TF_time, PT_time],color=['blue','red'])
plt.ylabel('Training Time',fontsize=15)
plt.xticks([0,1],['TensorFlow','PyTorch'],fontsize=15)
```

```python
resTF_time,resTF_acc=np.load('TF+Keras_resnet.npy')
resPT_time,resPT_acc=np.load('PT+fastai.npy')
```

```python
plt.figure(figsize=(7,7))
plt.bar([0,1], [resTF_acc, resPT_acc],color=['blue','red'])
plt.ylabel('Accuracy',fontsize=16)
plt.xticks([0,1],['TensorFlow','PyTorch'],fontsize=15)
```

```python
plt.figure(figsize=(7,7))
plt.bar([0,1], [resTF_time, resPT_time],color=['blue','red'])
plt.ylabel('Training Time',fontsize=15)
plt.xticks([0,1],['TensorFlow','PyTorch'],fontsize=15)
```
